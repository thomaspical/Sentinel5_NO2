# Etude journali√®re de la concentration de NO2

```python
import pandas as pd
import numpy as np
import os

import plotly.graph_objects as go
import plotly.express as px
```

```python
#Config
path="datas_sentinel5/"
```

```python
df_NO2=pd.DataFrame()

ls=[i for i in os.listdir(path+"archives") if os.path.isfile(os.path.join(path+"archives",i)) and 'archived' in i]
for elem in ls:
    df_NO2_new=pd.read_csv(path+"archives/"+elem, index_col=0)
    df_NO2_new=df_NO2_new.replace("","Undefined")
    df_NO2=pd.concat([df_NO2, df_NO2_new], axis=0)
    
df_NO2=df_NO2.reset_index(drop=True)
#Re-arranging date
df_NO2=df_NO2.rename(columns={"day_mean":"day"})
df_NO2["date"]=pd.to_datetime(df_NO2[["year","month","day"]])
df_NO2=df_NO2.drop(columns=["year","month","day"])
```

```python
#df_NO2.loc[:,["cc_departement","cc_region","cc_ville"]]=df_NO2.loc[:,["cc_departement","cc_region","cc_ville"]].fillna("Undefined")
```

```python
df_NO2["calcul"]=df_NO2.NO2*df_NO2.counter
```

```python
#Evolution of pollution for some countries
df_NO2_group=df_NO2.groupby(["cc_pays",df_NO2.date.dt.year,df_NO2.date.dt.month,df_NO2.date.dt.day]).agg({'counter':'sum','calcul':'sum'})
df_NO2_group["NO2"]=df_NO2_group.calcul/df_NO2_group.counter
```

```python
df_NO2_group["date"]=pd.to_datetime(df_NO2_group.index.get_level_values(1).astype(str) + df_NO2_group.index.get_level_values(2).astype(str) + df_NO2_group.index.get_level_values(3).astype(str), format="%Y%m%d")
df_NO2_group["cc_pays"]=df_NO2_group.index.get_level_values(0).astype(str)
```

```python
ls=["FR","CN","IT","JP","DE","IN"]
fig = px.line(df_NO2_group[[elem in ls for elem in df_NO2_group.cc_pays]], y="NO2", x="date", color="cc_pays", title="NO2 Evolution for countries", render_mode="webgl")
fig.show()
```

```python
ls=["FR","CN","IT","JP","DE","IN"]
fig = px.line(df_NO2_group[[elem in ls for elem in df_NO2_group.cc_pays]], y="counter", x="date", color="cc_pays", title="Observations evolution for countries", render_mode="webgl")
fig.show()
```

```python
fig = px.box(df_NO2_group[[elem in ls for elem in df_NO2_group.cc_pays]], x="cc_pays", y="counter", color="cc_pays",points="all")
fig.show()
```

```python
#Pollution evolution for french regions
df_NO2_group=df_NO2.groupby(["cc_region",df_NO2.date.dt.year,df_NO2.date.dt.month,df_NO2.date.dt.day]).agg({'counter':'sum','calcul':'sum'})
df_NO2_group["NO2"]=df_NO2_group.calcul/df_NO2_group.counter
```

```python
df_NO2_group["date"]=pd.to_datetime(df_NO2_group.index.get_level_values(1).astype(str) + df_NO2_group.index.get_level_values(2).astype(str) + df_NO2_group.index.get_level_values(3).astype(str), format="%Y%m%d")
df_NO2_group["cc_region"]=df_NO2_group.index.get_level_values(0).astype(str)
```

```python
ls=["Ile-de-France","Nord-Pas-de-Calais","Pyongyang","Hubei", "New York","Beijing", "Berlin"]
fig = px.line(df_NO2_group[[elem in ls for elem in df_NO2_group.cc_region]], y="NO2", x="date", color="cc_region", title="NO2 Evolution for regions", render_mode="webgl")
fig.show()
```

```python
fig = px.line(df_NO2_group[[elem in ls for elem in df_NO2_group.cc_region]], y="counter", x="date", color="cc_region", title="Observations evolution for regions", render_mode="webgl")
fig.show()
```

```python
fig = px.box(df_NO2_group[[elem in ls for elem in df_NO2_group.cc_region]], x="cc_region", y="counter", color="cc_region",points="all")
fig.show()
```

```python
#Worldwide sans les oceans 
df_NO2_group=df_NO2[(df_NO2.cc_pays != "Undefined") & (df_NO2.cc_departement != "Undefined") & (df_NO2.cc_region != "Undefined")].groupby([df_NO2.date.dt.year,df_NO2.date.dt.month,df_NO2.date.dt.day]).agg({'counter':'sum','calcul':'sum'})
df_NO2_group["NO2"]=df_NO2_group.calcul/df_NO2_group.counter
df_NO2_group["date"]=pd.to_datetime(df_NO2_group.index.get_level_values(0).astype(str) + df_NO2_group.index.get_level_values(1).astype(str) + df_NO2_group.index.get_level_values(2).astype(str), format="%Y%m%d")
```

```python
fig = px.line(df_NO2_group, y="NO2", x="date", render_mode="webgl", title="Global average NO2 level outside the oceans")
fig.show()
```

```python
fig = px.line(df_NO2_group, y="counter", x="date", render_mode="webgl", title="Observations evolutions at defined places")
fig.show()
```
